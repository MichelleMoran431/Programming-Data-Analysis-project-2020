### Variable 1 : AGE PROFILE

### Generate random values of age using Numpy.random function

```python
# Generate age between 20 and 45 years old using the Percent probabilities from survey
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()

np.random.seed(42)

AgeProfile =20,25,30,35,40
AgeProfile = np.random.choice(AgeProfile, 500, p=[0.048, 0.141, 0.366, 0.358, 0.087]) 
AgeProfile

# Average age of the sample
np.mean(AgeProfile)
#np.median(AgeProfile)S
```

```python
# Average age of the sample
import numpy as np
np.mean(AgeProfile)
#np.median(AgeProfile)
```

#### The mean age is 31.22  which is in line the national average of 32.5 years

```python
#Plot the age using a histogram to visualize the distribution of this variable
# Histogram

plt.figure(figsize=(12,8), facecolor='w', edgecolor='k', clear='True')
bins=np.linspace (20,45,50) #linspace returns evenly spaced numbers over a specified interval
plt.hist(AgeProfile, bins)# adding distribution samples 
plt.xlabel('Age in years')
plt.ylabel('Frequency')
plt.title('Age of Mothers who took the Survey')

plt.show()
```

```python
#Boxplot
import matplotlib.pyplot as plt
plt.boxplot(AgeProfile, showmeans=True)
plt.show()
```

##### From the histogram above , we can see the peak ages for women is between 30 and 35. Also that the distribution is a normal , as from the boxplot , shows the mean and median are similar
